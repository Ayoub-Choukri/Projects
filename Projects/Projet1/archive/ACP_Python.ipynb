```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv("./eauxTP.csv",sep= ' ',header=0,index_col=1)


data = data.drop(columns='Nom')

df= data.iloc[:,4:]
p= df.shape[1]
df
```

```python


# Créez une instance de PCA
pca = PCA()

# Appliquez PCA aux données du DataFrame (supposons que "df" contient vos données)
pca.fit(df)

# Obtenez les ratios de variance cumulée inversés
l=np.cumsum(pca.explained_variance_ratio_)

plt.bar(range(1,len(l)+1),l)
```

```python
# Créez une instance de PCA
pca = PCA()

# Appliquez PCA aux données du DataFrame (supposons que "df" contient vos données)
pca.fit(df)

n = df.shape[0]  # Nombre d'individus à tirer aléatoirement

# Obtenez le nombre total d'individus dans le DataFrame
nombre_total_individus = df.shape[0]

# Générez n indices aléatoires entre 0 et le nombre total d'individus
indices_aleatoires = np.random.choice(nombre_total_individus, n, replace=False)

# Sélectionnez les données correspondant aux indices aléatoires
donnees_aleatoires = df.iloc[indices_aleatoires]

# Transformez les données avec PCA
C = pca.transform(donnees_aleatoires)

names = donnees_aleatoires.index

plt.figure(figsize=(10,8))
scatter=plt.scatter(C[:, 0], C[:, 1],c=donnees_aleatoires['Cal'])

# Ajoutez une ligne en pointillé pour l'axe y=0
plt.axvline(x=0, color='gray', linestyle='--')

# Ajoutez une ligne en pointillé pour l'axe x=0
plt.axhline(y=0, color='gray', linestyle='--')

for x, y, nom in zip(C[:, 0], C[:, 1], names):
    plt.text(x + 0.02, y, nom, fontsize=8)
cbar = plt.colorbar(scatter)
plt.show()
```
